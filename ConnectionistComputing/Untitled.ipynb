```python
import csv
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/judy/Desktop/movies-sample.csv',usecols=[2])
df
```

```python
df['AB_split'] = df['type'].str.split('|')
df
```

```python
list=[]
dic={}
for index,row in df.iterrows():
    text=row['type']
    genre=text.split('|')
    for i in genre:
        if i in dic:
            dic[i]+=1
        else:
            dic[i]=0
dic
```

```python
import operator
dic = dict(sorted(dic.items(), key=operator.itemgetter(1),reverse=True))
dic
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(35,20))
plt.rcParams.update({'font.size': 22})
plt.xlabel("X axis label")
plt.ylabel("Y axis label")
plt.xticks(rotation=45)
plt.bar((dic.keys()), dic.values(), color='g')
# plt.figure(figsize=(30,30 。))
# plt.xticks([0,np.pi/2,np.pi,3*np.pi/2,2*np.pi],['0',r'$\frac{\pi}{2}$',r'$\pi$',r'$\frac{3\pi}{2}$',r'$2\pi$'], rotation=90)# 第一个参数是值，第二个参数是对应的显示效果(若无传入则默认直接显示原始数据)，第三个参数是标签旋转角度


plt.show()
```
